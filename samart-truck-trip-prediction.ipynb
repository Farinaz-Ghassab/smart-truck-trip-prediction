```python
df = pd.read_excel(r"C:\Users\fArInAz\OneDrive\Документы\transport_data.xlsx")

X = df[['distance_km' , 'weight_ton']]
y_fuel = df['fuel_liter']
y_price = df['price_toman']
```

```python
from sklearn.ensemble import RandomForestRegressor

fuel_model = RandomForestRegressor(n_estimators=50, random_state=42)
price_model = RandomForestRegressor(n_estimators=50, random_state=42)

fuel_model.fit(X, y_fuel)
price_model.fit(X, y_price)
```

```python
def predict_truck_trip(distance_km, weight_ton):
    input_data = pd.DataFrame([[distance_km, weight_ton]], columns=['distance_km', 'weight_ton'])
    
    predicted_fuel = fuel_model.predict(input_data)[0]
    predicted_price = price_model.predict(input_data)[0]
    return predicted_fuel, predicted_price
```

```python
fuel, price = predict_truck_trip(620, 5)
print(fuel, price)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
```

```python
X = df[['distance_km' , 'weight_ton']]
y_fuel = df['fuel_liter']
y_price = df['price_toman'] 
               
X_train, X_test, y_fuel_train, y_fuel_test = train_test_split(X, y_fuel, test_size=0.2, random_state=42)
_, _, y_price_train, y_price_test = train_test_split(X, y_price, test_size=0.2, random_state=42)
```

```python
model_fuel = RandomForestRegressor(n_estimators=50, random_state=42)
model_price = RandomForestRegressor(n_estimators=50, random_state=42)

model_fuel.fit(X_train, y_fuel_train)
model_price.fit(X_train, y_price_train)
```

```python
fuel_pred = model_fuel.predict(X_test)
price_pred = model_price.predict(X_test)
```

```python
fuel_mae = mean_absolute_error(y_fuel_test, fuel_pred)
fuel_rmse = np.sqrt(mean_squared_error(y_fuel_test, fuel_pred))
fuel_r2 = r2_score(y_fuel_test, fuel_pred)

price_man = mean_absolute_error(y_price_test, price_pred)
price_rmse = np.sqrt(mean_squared_error(y_price_test, price_pred))
price_r2 = r2_score(y_price_test, price_pred)

print("Fuel Model Accuracy")
print("MAE:", round(fuel_mae, 2))
print("RMSE:", round(fuel_rmse, 2))
print("R2:", round(fuel_r2, 3))

print("Price Model Accuracy")
print("MAE:", int(price_man))
print("RMSE:", int(price_rmse))
print("R2:", round(price_r2, 3))
```

```python
def get_distance_from_excel(origin_city, destination_city):
    row = df[(df['origin_city'] == origin_city) & (df['destination_city'] == destination_city)]
    return float(row.iloc[0]['distance_km'])
```

```python
def predict_trip_from_excel(origin_city, destination_city, weight_ton):
    distance = get_distance_from_excel(origin_city, destination_city)
    fuel, price = predict_truck_trip(distance, weight_ton)
    return { "origin": origin_city, "destination": destination_city, "distance_km": distance, " fuel_liter": round(fuel, 1), "price_toman": int(price)}
```
